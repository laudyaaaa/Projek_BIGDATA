EDA DAN EKSPLORASI

```python
# ============================================================
# FUNGSI EDA REUSABLE
# ============================================================

def explore_dataframe(df, name):
    """
    Menampilkan ringkasan lengkap DataFrame:
    - Shape, dtypes, null values, statistik deskriptif, dan sampel data
    """
    print(f"\n{'='*60}")
    print(f"EKSPLORASI DATA: {name.upper()}")
    print(f"{'='*60}")

    # 1. Shape
    print(f"Dimensi DataFrame:")
    print(f"   → {df.shape[0]:,} baris × {df.shape[1]} kolom")

    # 2. Tipe data
    print(f"Tipe Data (dtypes):")
    dtype_info = df.dtypes.reset_index()
    dtype_info.columns = ['Kolom', 'Dtype']
    print(dtype_info.to_string(index=False))

    # 3. Missing values
    print(f"Missing Values:")
    null_info = df.isnull().sum().reset_index()
    null_info.columns = ['Kolom', 'Jumlah Null']
    null_info['Persentase (%)'] = (null_info['Jumlah Null'] / len(df) * 100).round(2)
    print(null_info.to_string(index=False))

    # 4. Duplikasi
    dup_count = df.duplicated().sum()
    print(f"Data Duplikat: {dup_count:,} baris ({dup_count/len(df)*100:.2f}%)")

    # 5. Distribusi rating
    if 'score' in df.columns:
        print(f"Distribusi Rating (score):")
        rating_dist = df['score'].value_counts().sort_index()
        for star, count in rating_dist.items():
            bar = '★' * star + '☆' * (5 - star)
            pct = count / len(df) * 100
            print(f"   {bar} ({star}): {count:,} ({pct:.1f}%)")
        print(f"   Rata-rata rating: {df['score'].mean():.3f}")
        print(f"   Median rating   : {df['score'].median():.1f}")

    # 6. Rentang waktu
    if 'at' in df.columns:
        df['at'] = pd.to_datetime(df['at'])
        print(f"Rentang Waktu Review:")
        print(f"   Dari  : {df['at'].min().strftime('%d %B %Y')}")
        print(f"   Hingga: {df['at'].max().strftime('%d %B %Y')}")
        print(f"   Durasi: {(df['at'].max() - df['at'].min()).days} hari")

    # 7. Panjang teks review
    if 'content' in df.columns:
        df['text_length'] = df['content'].astype(str).apply(len)
        df['word_count'] = df['content'].astype(str).apply(lambda x: len(x.split()))
        print(f"Statistik Panjang Teks Review:")
        print(f"   Karakter - Min: {df['text_length'].min()}, Max: {df['text_length'].max()}, Rata-rata: {df['text_length'].mean():.1f}")
        print(f"   Kata     - Min: {df['word_count'].min()}, Max: {df['word_count'].max()}, Rata-rata: {df['word_count'].mean():.1f}")

    # 8. Sampel data
    print(f"Sampel 5 Data Pertama:")
    display_cols = ['userName', 'content', 'score', 'at'] if all(c in df.columns for c in ['userName', 'content', 'score', 'at']) else df.columns[:5]
    print(df[display_cols].head().to_string())
```

HYBRID LABELING


```python
# =========================================
# IMPORT LIBRARY
# =========================================

import pandas as pd
import numpy as np

# =========================================
# LEXICON POSITIF & NEGATIF
# =========================================

positive_words = [
    'bagus', 'cepat', 'murah', 'puas',
    'baik', 'mantap', 'sesuai', 'recommended',
    'aman', 'top', 'keren', 'suka'
]

negative_words = [
    'lama', 'rusak', 'buruk', 'kecewa',
    'jelek', 'parah', 'mahal', 'lambat',
    'benci', 'cacat', 'zonk', 'error'
]

# =========================================
# LABEL DARI RATING
# =========================================

def label_from_rating(rating):

    if rating >= 4:
        return 'positif'

    elif rating <= 2:
        return 'negatif'

    else:
        return 'netral'

# =========================================
# LABEL DARI KOMENTAR
# =========================================

def label_from_comment(text):

    text = str(text).lower()

    pos_score = 0
    neg_score = 0

    # Hitung kata positif
    for word in positive_words:
        if word in text:
            pos_score += 1

    # Hitung kata negatif
    for word in negative_words:
        if word in text:
            neg_score += 1

    # Tentukan label
    if pos_score > neg_score:
        return 'positif'

    elif neg_score > pos_score:
        return 'negatif'

    else:
        return 'unknown'

# =========================================
# HYBRID LABELING FUNCTION
# =========================================

def hybrid_label(row):

    rating_label = row['rating_label']
    comment_label = row['comment_label']

    # Rating 3 → netral
    if rating_label == 'netral':
        return 'netral'

    # Cocok
    elif rating_label == comment_label:
        return rating_label

    # Unknown → pakai rating
    elif comment_label == 'unknown':
        return rating_label

    # Tidak cocok
    else:
        return 'mismatch'

# =========================================
# TERAPKAN HYBRID LABELING KE SEMUA PLATFORM
# =========================================

for df_platform, name in [
    (df_shopee, 'Shopee'),
    (df_lazada, 'Lazada'),
    (df_tokopedia, 'Tokopedia')
]:
    df_platform['rating_label'] = df_platform['score'].apply(label_from_rating)
    df_platform['comment_label'] = df_platform['content'].apply(label_from_comment)
    df_platform['hybrid_label'] = df_platform.apply(hybrid_label, axis=1)
    print(f"\nDistribusi Hybrid Label - {name}:")
    print(df_platform['hybrid_label'].value_counts())

# =========================================
# GABUNGKAN SEMUA PLATFORM (untuk dipakai di section selanjutnya)
# =========================================

df_shopee['platform'] = 'Shopee'
df_lazada['platform'] = 'Lazada'
df_tokopedia['platform'] = 'Tokopedia'

df = pd.concat([df_shopee, df_lazada, df_tokopedia], ignore_index=True)

print("\nTotal data gabungan:", df.shape)
```

```python
# =========================================================
# ANALISIS KONSISTENSI RATING VS KOMENTAR
# SHOPEE - LAZADA - TOKOPEDIA
# =========================================================

# =========================================================
# IMPORT LIBRARY
# =========================================================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# =========================================================
# DATAFRAME
# =========================================================
# Pastikan dataframe bernama:
# df
#
# Dan memiliki kolom:
# - review
# - rating
# - platform

# =========================================================
# LABEL DARI RATING
# =========================================================
# Rating:
# 1-2 = negatif
# 4-5 = positif
# 3 = netral

def label_from_rating(rating):

    if rating >= 4:
        return 'positif'

    elif rating <= 2:
        return 'negatif'

    else:
        return 'netral'

# =========================================================
# MEMBUAT LABEL RATING
# =========================================================

df['rating_label'] = df['score'].apply(label_from_rating)

# =========================================================
# LEXICON POSITIF
# =========================================================

positive_words = [

    'bagus',
    'cepat',
    'murah',
    'puas',
    'baik',
    'mantap',
    'sesuai',
    'recommended',
    'aman',
    'keren',
    'suka'

]

# =========================================================
# LEXICON NEGATIF
# =========================================================

negative_words = [

    'lama',
    'rusak',
    'buruk',
    'kecewa',
    'jelek',
    'parah',
    'mahal',
    'lambat',
    'zonk',
    'cacat',
    'error'

]

# =========================================================
# LABEL DARI KOMENTAR
# =========================================================

def label_from_comment(text):

    text = str(text).lower()

    pos_score = 0
    neg_score = 0

    # Hitung kata positif
    for word in positive_words:

        if word in text:
            pos_score += 1

    # Hitung kata negatif
    for word in negative_words:

        if word in text:
            neg_score += 1

    # Menentukan label komentar
    if pos_score > neg_score:

        return 'positif'

    elif neg_score > pos_score:

        return 'negatif'

    else:

        return 'unknown'

# =========================================================
# MEMBUAT LABEL KOMENTAR
# =========================================================

df['comment_label'] = df['content'].apply(label_from_comment)

# =========================================================
# ANALISIS KONSISTENSI
# =========================================================

def consistency_check(row):

    rating_label = row['rating_label']
    comment_label = row['comment_label']

    # Jika rating 3
    if rating_label == 'netral':

        return 'unknown'

    # Jika komentar tidak terbaca
    elif comment_label == 'unknown':

        return 'unknown'

    # Jika rating dan komentar sama
    elif rating_label == comment_label:

        return 'konsisten'

    # Jika berbeda
    else:

        return 'mismatch'

# =========================================================
# MEMBUAT KOLOM KONSISTENSI
# =========================================================

df['consistency'] = df.apply(
    consistency_check,
    axis=1
)

# =========================================================
# HASIL ANALISIS
# =========================================================

print(df[
    [
        'content',
        'score',
        'platform',
        'rating_label',
        'comment_label',
        'consistency'
    ]
].head())

# =========================================================
# DISTRIBUSI KONSISTENSI
# =========================================================

print("\nDistribusi Konsistensi:")
print(df['consistency'].value_counts())

# =========================================================
# PERSENTASE KONSISTENSI
# =========================================================

consistency_percentage = (

    df['consistency']
    .value_counts(normalize=True)
    .mul(100)

)

print("\nPersentase Konsistensi:")
print(consistency_percentage)

# =========================================================
# KONSISTENSI PER PLATFORM
# =========================================================

platform_consistency = (

    df.groupby('platform')['consistency']
    .value_counts(normalize=True)
    .mul(100)
    .rename('persentase')
    .reset_index()

)

print("\nKonsistensi per Platform:")
print(platform_consistency)

# =========================================================
# VISUALISASI KONSISTENSI
# =========================================================

plt.figure(figsize=(8,5))

sns.countplot(
    x='consistency',
    data=df
)

plt.title('Analisis Konsistensi Rating vs Komentar')
plt.xlabel('Kategori')
plt.ylabel('Jumlah Data')

plt.show()

# =========================================================
# VISUALISASI KONSISTENSI PER PLATFORM
# =========================================================

plt.figure(figsize=(10,6))

sns.countplot(
    x='platform',
    hue='consistency',
    data=df
)

plt.title('Konsistensi Rating-Komenter per Platform')
plt.xlabel('Platform')
plt.ylabel('Jumlah Data')

plt.show()
```

```python
normalisasi = {
    'gak': 'tidak', 'ga': 'tidak', 'gk': 'tidak', 'nggak': 'tidak',
    'ngga': 'tidak', 'enggak': 'tidak', 'udah': 'sudah', 'udh': 'sudah',
    'sdh': 'sudah', 'blm': 'belum', 'belom': 'belum', 'kmrn': 'kemarin',
    'skrg': 'sekarang', 'skrang': 'sekarang', 'skg': 'sekarang',
    'lg': 'lagi', 'lgi': 'lagi', 'sm': 'sama', 'ama': 'sama',
    'bgt': 'banget', 'bget': 'banget', 'bingit': 'banget',
    'emg': 'memang', 'emang': 'memang', 'gmn': 'bagaimana',
    'gmana': 'bagaimana', 'gimana': 'bagaimana', 'knp': 'kenapa',
    'krn': 'karena', 'karna': 'karena', 'krna': 'karena',
    'jg': 'juga', 'jga': 'juga', 'yg': 'yang', 'yng': 'yang',
    'dgn': 'dengan', 'dg': 'dengan', 'dr': 'dari', 'utk': 'untuk',
    'buat': 'untuk', 'sy': 'saya', 'gue': 'saya', 'gw': 'saya',
    'ane': 'saya', 'aku': 'saya', 'lu': 'kamu', 'lo': 'kamu',
    'elo': 'kamu', 'klo': 'kalau', 'klu': 'kalau', 'kl': 'kalau',
    'kalo': 'kalau', 'abis': 'habis', 'abiss': 'habis',
    'msh': 'masih', 'tp': 'tapi', 'tpi': 'tapi', 'tetapi': 'tapi',
    'td': 'tadi', 'hrs': 'harus', 'hrus': 'harus', 'brg': 'barang',
    'brng': 'barang', 'pnjual': 'penjual', 'pmbeli': 'pembeli',
    'bayar': 'bayar', 'byr': 'bayar', 'harga': 'harga', 'hrg': 'harga',
    'mau': 'mau', 'mw': 'mau', 'bisa': 'bisa', 'bs': 'bisa',
    'tdk': 'tidak', 'tak': 'tidak', 'tdk': 'tidak',
    'mantap': 'mantap', 'mantul': 'mantap', 'keren': 'bagus',
    'jelek': 'buruk', 'jeble': 'buruk', 'parah': 'buruk',
    'oke': 'baik', 'ok': 'baik', 'oks': 'baik',
    'lmbt': 'lambat', 'lemot': 'lambat', 'lelet': 'lambat',
    'cpt': 'cepat', 'cepet': 'cepat', 'cepatt': 'cepat',
    'murah': 'murah', 'mrah': 'murah', 'mahal': 'mahal',
    'tlg': 'tolong', 'tolong': 'tolong', 'minta': 'minta',
    'mohon': 'mohon', 'mhn': 'mohon', 'pd': 'pada', 'bener²': 'benar-benar', 'bagu': 'bagus'
}

def normalize_word(text):
    words = text.split()
    new_words = []
    for w in words:
        new_words.append(normalisasi[w] if w in normalisasi else w)
    return " ".join(new_words)
```

```python
stopwords_id = set(stopwords.words('indonesian'))
stopwords_extra = {
    'app', 'aplikasi', 'shopee', 'lazada', 'nya', 'nih', 'sih', 'deh',
    'dong', 'lah', 'kah', 'pun', 'tuh', 'gitu', 'gini', 'aja', 'ajalah',
    'yah', 'ya', 'kan', 'mah', 'nah', 'ih', 'ah', 'oh', 'eh', 'wah',
    'haha', 'hihi', 'wkwk', 'kwkw', 'hehe', 'hoho', 'huhu',
    'nan', 'ku', 'mu', 'kamu', 'kami', 'kalian', 'mereka',
    'ini', 'itu', 'tersebut', 'tadi'
}
stopwords_id.update(stopwords_extra)
```

```python
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"http\S+|www\S+", "", text)
    text = re.sub(r"@\w+","", text)
    text = re.sub(r"#\w+","", text)
    text = re.sub(r"\d+","", text)
    text = text.translate(str.maketrans("", "", string.punctuation))
    text = re.sub(r"\s+", " ", text).strip()
    text = normalize_word(text)
    words = [w for w in text.split() if w not in stopwords_id]
    return " ".join(words)
```

```python
# =========================================================
# FEATURE EXTRACTION
# TF-IDF VECTORIZER
# =========================================================

# =========================================================
# IMPORT LIBRARY
# =========================================================

from sklearn.feature_extraction.text import TfidfVectorizer

# =========================================================
# KONDISI DATA
# =========================================================
# Pastikan dataframe:
# df_filtered
#
# Sudah memiliki kolom:
# - clean_review
# - label
#
# clean_review berasal dari preprocessing teks

# Add the clean_review column to df_filtered
df_filtered['clean_review'] = df_filtered['content'].apply(clean_text)

# =========================================================
# CEK DATA
# =========================================================

print(df_filtered[['clean_review', 'label']].head())

# =========================================================
# MEMISAHKAN FITUR DAN LABEL
# =========================================================

X = df_filtered['clean_review']

y = df_filtered['label']

# =========================================================
# TF-IDF VECTORIZER
# =========================================================

tfidf = TfidfVectorizer(

    max_features=5000,     # maksimal jumlah fitur
    ngram_range=(1,2),     # unigram + bigram
    min_df=5,              # minimal muncul 5x
    max_df=0.9             # maksimal 90%

)

# =========================================================
# FIT DAN TRANSFORM
# =========================================================

X_tfidf = tfidf.fit_transform(X)

# =========================================================
# CEK HASIL TF-IDF
# =========================================================

print("\nShape TF-IDF:")
print(X_tfidf.shape)

# =========================================================
# MELIHAT FITUR/KATA
# =========================================================

feature_names = tfidf.get_feature_names_out()

print("\nContoh Feature:")
print(feature_names[:20])

# =========================================================
# MELIHAT TF-IDF MATRIX
# =========================================================

tfidf_df = pd.DataFrame(

    X_tfidf.toarray(),
    columns=feature_names

)

print("\nContoh TF-IDF Matrix:")
print(tfidf_df.head())

# =========================================================
# TOP KATA PALING SERING
# =========================================================

word_scores = np.asarray(X_tfidf.sum(axis=0)).ravel()

top_words = pd.DataFrame({

    'word': feature_names,
    'score': word_scores

})

top_words = top_words.sort_values(
    by='score',
    ascending=False
)

print("\nTop 20 Kata TF-IDF:")
print(top_words.head(20))

# =========================================================
# VISUALISASI TOP WORD
# =========================================================

import matplotlib.pyplot as plt

top_10 = top_words.head(10)

plt.figure(figsize=(10,5))

plt.bar(
    top_10['word'],
    top_10['score']
)

plt.title('Top 10 Kata TF-IDF')
plt.xlabel('Kata')
plt.ylabel('Score')

plt.xticks(rotation=45)

plt.show()
```

```python
from collections import Counter

# ============================================================
# ANALISIS FREKUENSI KATA
# ============================================================

def get_top_words(df, text_col='clean', n=30, sentiment=None):
    """
    Menghitung frekuensi kata dari kolom teks.
    Bisa difilter berdasarkan sentimen.
    """
    if sentiment:
        filtered = df[df['sentimen'] == sentiment][text_col]
    else:
        filtered = df[text_col]

    # Gabungkan semua teks dan split
    all_words = ' '.join(filtered.dropna()).split()
    word_freq = Counter(all_words)

    return word_freq.most_common(n)

# Top kata keseluruhan
print("📊 TOP 20 KATA - SHOPEE:")
top_shopee = get_top_words(df_shopee_fe, text_col='clean', n=20)
for i, (word, freq) in enumerate(top_shopee, 1):
    bar = '█' * (freq // max(top_shopee[0][1] // 30, 1))
    print(f"  {i:2}. {word:20} {bar} {freq:,}")

print("\n📊 TOP 20 KATA - LAZADA:")
top_lazada = get_top_words(df_lazada_fe, text_col='clean', n=20)
for i, (word, freq) in enumerate(top_lazada, 1):
    bar = '█' * (freq // max(top_lazada[0][1] // 30, 1))
    print(f"  {i:2}. {word:20} {bar} {freq:,}")

print("\n📊 TOP 20 KATA - TOKOPEDIA:")
top_tokopedia = get_top_words(df_tokopedia_fe, text_col='clean', n=20)
for i, (word, freq) in enumerate(top_tokopedia, 1):
    bar = '█' * (freq // max(top_tokopedia[0][1] // 30, 1))
    print(f"  {i:2}. {word:20} {bar} {freq:,}")
```

```python
# ============================================================
# N-GRAM ANALYSIS (Bigram & Trigram)
# ============================================================
# N-gram menangkap frasa yang sering muncul bersamaan
# Bigram = 2 kata, Trigram = 3 kata
# Contoh: 'pengiriman lambat', 'tidak bisa login', 'harga sangat murah'

from sklearn.feature_extraction.text import CountVectorizer

def get_ngrams(df, text_col='clean', n=2, top_k=20, sentiment=None):
    """
    Ekstrak n-gram paling sering muncul.
    Gunakan text_nostem untuk bigram lebih natural dibaca.
    """
    if sentiment:
        texts = df[df['sentimen'] == sentiment][text_col].dropna()
    else:
        texts = df[text_col].dropna()

    vectorizer = CountVectorizer(
        ngram_range=(n, n),
        max_features=top_k * 3,
        min_df=5           # Muncul minimal di 5 review berbeda
    )

    try:
        X = vectorizer.fit_transform(texts)
        vocab = vectorizer.get_feature_names_out()
        freqs = np.asarray(X.sum(axis=0)).flatten()
        ngram_freq = sorted(zip(vocab, freqs), key=lambda x: x[1], reverse=True)
        return ngram_freq[:top_k]
    except Exception as e:
        print(f"Error processing n-grams: {e}")
        return []

fig, axes = plt.subplots(3, 2, figsize=(16, 15)) # Changed to 3 rows, 2 columns

configs = [
    (df_shopee_fe, 'Shopee', 'Positif', '#27ae60', 2),
    (df_shopee_fe, 'Shopee', 'Negatif', '#c0392b', 2),
    (df_lazada_fe, 'Lazada', 'Positif', '#27ae60', 2),
    (df_lazada_fe, 'Lazada', 'Negatif', '#c0392b', 2),
    (df_tokopedia_fe, 'Tokopedia', 'Positif', '#27ae60', 2), # Added Tokopedia
    (df_tokopedia_fe, 'Tokopedia', 'Negatif', '#c0392b', 2)  # Added Tokopedia
]

for ax, (df, platform, sentiment, color, n) in zip(axes.flat, configs):
    ngrams = get_ngrams(df, n=n, top_k=12, sentiment=sentiment)

    if ngrams:
        labels, values = zip(*ngrams)
        bars = ax.barh(range(len(labels)), values, color=color, alpha=0.8)
        ax.set_yticks(range(len(labels)))
        ax.set_yticklabels(labels, fontsize=9)
        ax.invert_yaxis()
        ax.set_title(f'{platform} – Top Bigram ({sentiment})', fontweight='bold')
        ax.set_xlabel('Frekuensi')

plt.suptitle('🔗 Analisis Bigram per Sentimen – Shopee vs Lazada vs Tokopedia',
             fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('ngram_analysis.png', bbox_inches='tight', dpi=150)
plt.show()
```

```python
aspect_keywords = {
    'Pengiriman': ['kirim', 'pengiriman', 'kurir', 'paket', 'tiba', 'sampai', 'lambat', 'cepat',
                   'logistik', 'ekspedisi', 'jne', 'jnt', 'sicepat', 'anteraja', 'terlambat'],
    'Pembayaran': ['bayar', 'pembayaran', 'transfer', 'ovo', 'gopay', 'dana', 'qris', 'cod',
                   'cicilan', 'kredit', 'refund', 'uang', 'saldo', 'tagihan', 'kartu'],
    'Harga & Promo': ['harga', 'diskon', 'promo', 'voucher', 'cashback', 'murah', 'mahal',
                      'flash sale', 'gratis', 'ongkir', 'subsidi', 'kode'],
    'Produk & Penjual': ['produk', 'barang', 'penjual', 'seller', 'toko', 'kualitas', 'palsu',
                          'asli', 'original', 'rusak', 'baru', 'sesuai', 'deskripsi'],
    'Aplikasi & Teknis': ['aplikasi', 'error', 'bug', 'crash', 'login', 'loading', 'lambat',
                           'update', 'notifikasi', 'fitur', 'tampilan', 'ui', 'ux', 'lag'],
    'Customer Service': ['cs', 'customer service', 'layanan', 'komplain', 'respon', 'bantuan',
                          'solusi', 'hubungi', 'chat', 'admin', 'lapor', 'tidak direspon'],
    'Keamanan': ['aman', 'penipuan', 'scam', 'akun', 'hack', 'keamanan', 'privasi', 'data',
                  'verifikasi', 'otp', 'phishing']
}

def count_aspect_mentions(df, aspect_dict, text_col='content'):
    """
    Hitung berapa review yang menyebutkan setiap aspek,
    dipisah berdasarkan sentimen.
    """
    results = []
    texts_lower = df[text_col].str.lower()

    for aspect, keywords in aspect_dict.items():
        # Cari review yang mengandung keyword aspek tersebut
        mask = texts_lower.str.contains('|'.join(keywords), regex=True, na=False)
        aspect_df = df[mask]

        total = len(aspect_df)
        pos = (aspect_df['sentimen'] == 'Positif').sum()
        neg = (aspect_df['sentimen'] == 'Negatif').sum()
        avg_r = aspect_df['score'].mean() if total > 0 else 0

        results.append({
            'Aspek': aspect,
            'Total Mention': total,
            'Positif': pos,
            'Negatif': neg,
            'Avg Rating': round(avg_r, 2),
            'Pct Negatif': round(neg / total * 100, 1) if total > 0 else 0
        })

    return pd.DataFrame(results).sort_values('Total Mention', ascending=False)

aspect_shopee = count_aspect_mentions(df_shopee_fe, aspect_keywords)
aspect_lazada = count_aspect_mentions(df_lazada_fe, aspect_keywords)
aspect_tokopedia = count_aspect_mentions(df_tokopedia_fe, aspect_keywords)

print("📊 Analisis Aspek – SHOPEE:")
print(aspect_shopee.to_string(index=False))
print("\n📊 Analisis Aspek – LAZADA:")
print(aspect_lazada.to_string(index=False))
print("\n📊 Analisis Aspek – TOKOPEDIA:")
print(aspect_tokopedia.to_string(index=False))
```

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    classification_report
)
from sklearn.naive_bayes import MultinomialNB

import matplotlib.pyplot as plt
import seaborn as sns

# =========================================================
# MODEL TRAINING (Naive Bayes for Shopee as an example)
# =========================================================

# =========================================================
# TRAIN DAN EVALUASI NAIVE BAYES - SEMUA PLATFORM
# =========================================================

results_all = {}

for platform, X_train, X_test, y_train, y_test_data in [
    ('Shopee',    X_train_sh, X_test_sh, y_train_sh, y_test_sh),
    ('Lazada',    X_train_lz, X_test_lz, y_train_lz, y_test_lz),
    ('Tokopedia', X_train_tk, X_test_tk, y_train_tk, y_test_tk),
]:
    m = MultinomialNB()
    m.fit(X_train, y_train)
    preds = m.predict(X_test)
    results_all[platform] = {
        'model': m, 'y_test': y_test_data, 'y_pred': preds
    }
    print(f"\n✅ Model {platform} selesai dilatih")

# Gunakan Shopee sebagai contoh utama evaluasi detail
model = results_all['Shopee']['model']
y_pred = results_all['Shopee']['y_pred']
y_test = results_all['Shopee']['y_test']

# =========================================================
# HITUNG METRIK
# =========================================================

accuracy = accuracy_score(y_test, y_pred)

precision = precision_score(
    y_test,
    y_pred,
    average='weighted',
    zero_division=0 # Set zero_division to 0 to explicitly handle cases with no predicted samples
)

recall = recall_score(
    y_test,
    y_pred,
    average='weighted',
    zero_division=0 # Set zero_division to 0 to explicitly handle cases with no predicted samples
)

f1 = f1_score(
    y_test,
    y_pred,
    average='weighted',
    zero_division=0 # Set zero_division to 0 to explicitly handle cases with no predicted samples
)

# =========================================================
# HASIL EVALUASI
# =========================================================

print("HASIL EVALUASI MODEL\n")

print(f"Accuracy  : {accuracy:.4f}")
print(f"Precision : {precision:.4f}")
print(f"Recall    : {recall:.4f}")
print(f"F1-Score  : {f1:.4f}")

# =========================================================
# CLASSIFICATION REPORT
# =========================================================

print("\nClassification Report:\n")

print(
    classification_report(
        y_test,
        y_pred,
        zero_division=0 # Set zero_division to 0 for classification_report as well
    )
)

# =========================================================
# CONFUSION MATRIX
# =========================================================

cm = confusion_matrix(
    y_test,
    y_pred
)

plt.figure(figsize=(6,5))

sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Blues',
    xticklabels=sorted(set(y_test)),
    yticklabels=sorted(set(y_test))
)

plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('Actual Label')

plt.show()

# =========================================================
# VISUALISASI METRIK
# =========================================================

metrics = {
    'Accuracy': accuracy,
    'Precision': precision,
    'Recall': recall,
    'F1-Score': f1
}

plt.figure(figsize=(8,5))

bars = plt.bar(
    metrics.keys(),
    metrics.values()
)

for bar, val in zip(bars, metrics.values()):

    plt.text(
        bar.get_x() + bar.get_width()/2,
        val + 0.01,
        f'{val:.2f}',
        ha='center',
        fontweight='bold'
    )

plt.ylim(0,1.1)

plt.title('Evaluasi Model Naive Bayes')
plt.ylabel('Score')

plt.show()
```

```python
# =========================================================
# BIGRAM ANALYSIS
# =========================================================

from collections import Counter
from nltk.util import ngrams

# =========================================================
# FUNGSI BIGRAM
# =========================================================

def get_top_bigram(corpus, n=10):

    bigrams = []

    for text in corpus:

        tokens = str(text).split()

        bigrams.extend(
            list(ngrams(tokens, 2))
        )

    return Counter(bigrams).most_common(n)

# =========================================================
# BIGRAM PER PLATFORM
# =========================================================

platforms = [
    'Shopee',
    'Lazada',
    'Tokopedia'
]

for platform in platforms:

    print(f"\nTOP BIGRAM - {platform}")

    platform_text = df_filtered[
        df_filtered['platform'] == platform
    ]['clean_review']

    top_bigram = get_top_bigram(
        platform_text
    )

    for bg, count in top_bigram:

        print(f"{bg} : {count}")
```

```python
# =========================================================
# ISU UTAMA PER PLATFORM
# =========================================================

platforms = [
    'Shopee',
    'Lazada',
    'Tokopedia'
]

print("\nISU UTAMA PER PLATFORM\n")

for platform in platforms:

    platform_data = df_filtered[
        df_filtered['platform'] == platform
    ]

    positif = (
        platform_data['label'] == 'positif'
    ).mean() * 100

    negatif = (
        platform_data['label'] == 'negatif'
    ).mean() * 100

    print(f"{platform}")
    print(f"Positif : {positif:.2f}%")
    print(f"Negatif : {negatif:.2f}%")

    if negatif > 20:

        print("Isu utama: Banyak keluhan pengguna")

    else:

        print("Isu utama: Pengguna cenderung puas")

    print("-" * 40)
```

INSIGHT

```python
# =========================================================
# PLATFORM TERBAIK
# =========================================================

comparison = (

    df_filtered.groupby('platform')['label']
    .apply(lambda x: (x == 'positif').mean() * 100)

)

best_platform = comparison.idxmax()

best_score = comparison.max()

# =========================================================
# KESIMPULAN
# =========================================================

print("\nKESIMPULAN & INSIGHT\n")

print(
    f"Platform dengan sentimen positif tertinggi "
    f"adalah {best_platform} "
    f"dengan persentase positif sebesar "
    f"{best_score:.2f}%."
)

# =========================================================
# INSIGHT PER PLATFORM
# =========================================================

platforms = [
    'Shopee',
    'Lazada',
    'Tokopedia'
]

print("\nInsight:\n")

for platform in platforms:

    platform_data = df_filtered[
        df_filtered['platform'] == platform
    ]

    positive = (
        platform_data['label'] == 'positif'
    ).mean() * 100

    negative = (
        platform_data['label'] == 'negatif'
    ).mean() * 100

    print(
        f"{platform}: "
        f"Positif={positive:.2f}% | "
        f"Negatif={negative:.2f}%"
    )

# =========================================================
# LIMITASI PENELITIAN
# =========================================================

print("\nLimitasi Penelitian:")

print("- Label awal masih menggunakan rating pengguna")
print("- Lexicon sederhana belum memahami konteks kalimat")
print("- Bahasa slang dan singkatan masih sulit dipahami model")
```
