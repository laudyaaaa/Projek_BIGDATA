Preprocessing

```python
import re
import string
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
stopwords_id = set(stopwords.words('indonesian'))
stopwords_extra = {
    'yg','klo','kalo','ga','gak','iya','udah','aja','sih','nih','nya',
    'yah','tt','jd','jadi','dong','ko','kok','kayak','mah','lah','kan','pun',
    'deh','si','buat','dgn','tdk','gpp','banget','sangat','boleh','gitu',
    'gini','oke','ok','nya','kau','mu','kali'
}
stopwords_id.update(stopwords_extra)
```

```python
normalisasi = {
    "gak":"tidak", "ga":"tidak", "nggak":"tidak", "ngga":"tidak",
    "km":"kamu", "kmu":"kamu",
    "jgn":"jangan",
    "tp":"tapi", "tpi":"tapi",
    "dr":"dari", "pd":"pada",
    "krn":"karena",
    "brg":"barang",
    "ongkir":"ongkos_kirim",
    "packing":"paket",
    "pengiriman":"kirim",
    "ngirim":"kirim",
    "kirimnya":"kirim",
    "ngirim":"kirim",
    "ngepack":"packing",
    "spx":"kurir"
}

def normalize_word(text):
    words = text.split()
    new_words = []
    for w in words:
        new_words.append(normalisasi[w] if w in normalisasi else w)
    return " ".join(new_words)
```

```python
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"http\S+|www\S+", "", text)
    text = re.sub(r"@\w+","", text)
    text = re.sub(r"#\w+","", text)
    text = re.sub(r"\d+","", text)
    text = text.translate(str.maketrans("", "", string.punctuation))
    text = re.sub(r"\s+", " ", text).strip()
    text = normalize_word(text)
    words = [w for w in text.split() if w not in stopwords_id]
    return " ".join(words)
```

```python
df_shopee_fe = df_pengiriman_shopee.copy()
df_shopee_fe["clean"] = df_shopee_fe["content"].apply(clean_text)
```

```python
df_lazada_fe = df_pengiriman_lazada.copy()
df_lazada_fe["clean"] = df_lazada_fe["content"].apply(clean_text)
```

Feature Engineering

```python
def label_sentimen(nilai):
    if nilai >= 4:
        return 'Positif'
    elif nilai == 3:
        return 'Netral'
    else:
        return 'Negatif'

df_shopee_fe['sentimen'] = df_shopee_fe['score'].apply(label_sentimen)
df_lazada_fe['sentimen'] = df_lazada_fe['score'].apply(label_sentimen)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(
    max_features=5000,
    ngram_range=(1,2),     # unigram + bigram
    min_df=5,              # buang kata yang terlalu jarang
    max_df=0.9             # buang kata terlalu umum
)

tfidf_shopee = tfidf.fit_transform(df_shopee_fe['clean'])
tfidf_lazada = tfidf.fit_transform(df_lazada_fe['clean'])
```

```python
tfidf.get_feature_names_out()[:10]
```

```python
from sklearn.model_selection import train_test_split

X_train_sh, X_test_sh, y_train_sh, y_test_sh = train_test_split(
    tfidf_shopee, df_shopee_fe['sentimen'], test_size=0.2, random_state=42
)

X_train_lz, X_test_lz, y_train_lz, y_test_lz = train_test_split(
    tfidf_lazada, df_lazada_fe['sentimen'], test_size=0.2, random_state=42
)
```

Modeling

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
model_shopee = MultinomialNB()
model_shopee.fit(X_train_sh, y_train_sh)

y_pred_sh = model_shopee.predict(X_test_sh)
akurasi_shopee = accuracy_score(y_test_sh, y_pred_sh)

print("HASIL SHOPEE")
print("Akurasi Shopee:", round(akurasi_shopee, 3))
print("\nClassification Report Shopee:")
print(classification_report(y_test_sh, y_pred_sh))
```

```python
model_lazada = MultinomialNB()
model_lazada.fit(X_train_lz, y_train_lz)

y_pred_lz = model_lazada.predict(X_test_lz)
akurasi_lazada = accuracy_score(y_test_lz, y_pred_lz)

print("\nHASIL LAZADA")
print("Akurasi Lazada:", round(akurasi_lazada, 3))
print("\nClassification Report Lazada:")
print(classification_report(y_test_lz, y_pred_lz))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

from gensim.corpora import Dictionary
from gensim.models import CoherenceModel
```

```python
texts = df_shopee_fe['clean'].apply(lambda x: x.split()).tolist()
dictionary = Dictionary(texts)

coherence_values = []
topic_range = [2,3,4,5,6]

for k in topic_range:
    cv = CountVectorizer(max_features=2000)
    dtm = cv.fit_transform(df_shopee_fe['clean'])

    lda = LatentDirichletAllocation(
        n_components=k,
        random_state=42
    )
    lda.fit(dtm)

    topics = []
    feature_names = cv.get_feature_names_out()
    for topic in lda.components_:
        top_words = [feature_names[i] for i in topic.argsort()[-10:]]
        topics.append(top_words)

    cm = CoherenceModel(
        topics=topics,
        texts=texts,
        dictionary=dictionary,
        coherence='c_v'
    )
    coherence_values.append(cm.get_coherence())

list(zip(topic_range, coherence_values))
```

```python
k_final = 3
top_n = 10

sh_all = df_shopee_fe['clean']
sh_pos = df_shopee_fe[df_shopee_fe['sentimen']=="Positif"]['clean']
sh_neg = df_shopee_fe[df_shopee_fe['sentimen']=="Negatif"]['clean']

lz_all = df_lazada_fe['clean']
lz_pos = df_lazada_fe[df_lazada_fe['sentimen']=="Positif"]['clean']
lz_neg = df_lazada_fe[df_lazada_fe['sentimen']=="Negatif"]['clean']
```

```python
def fit_lda(text_series, k, top_n=10):
    cv = CountVectorizer(max_features=2000)
    dtm = cv.fit_transform(text_series)

    lda = LatentDirichletAllocation(
        n_components=k,
        random_state=42
    )
    lda.fit(dtm)

    feature_names = cv.get_feature_names_out()

    topics = []
    for topic in lda.components_:
        words = [feature_names[i] for i in topic.argsort()[-top_n:]]
        topics.append(words)

    doc_topic = lda.transform(dtm)
    props = doc_topic.mean(axis=0)

    return {
        'lda': lda,
        'dtm': dtm,
        'vectorizer': cv,
        'topics': topics,
        'props': props
    }
```

```python
res_lda = {}

res_lda['sh_pos'] = fit_lda(sh_pos, k_final, top_n)
res_lda['sh_neg'] = fit_lda(sh_neg, k_final, top_n)

res_lda['lz_pos'] = fit_lda(lz_pos, k_final, top_n)
res_lda['lz_neg'] = fit_lda(lz_neg, k_final, top_n)
```

```python
def print_topics(res, label):
    print(f"\nTOPIK {label}")
    for i, topic in enumerate(res['topics']):
        print(f"Topik {i+1}: {', '.join(topic)}")

print_topics(res_lda['sh_pos'], "Shopee Positif")
print_topics(res_lda['sh_neg'], "Shopee Negatif")
print_topics(res_lda['lz_pos'], "Lazada Positif")
print_topics(res_lda['lz_neg'], "Lazada Negatif")
```

```python
from wordcloud import WordCloud

def plot_wordcloud(res, label):
    all_words = []
    for topic in res['topics']:
        all_words.extend(topic)

    wordcloud = WordCloud(
        width=800,
        height=400,
        background_color='white'
    ).generate(" ".join(all_words))

    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(f'Word Cloud for {label}')
    plt.show()
```

```python
def plot_topic_comparison(res_sh, res_lz, label_sh, label_lz):
    topics = [f"Topik {i+1}" for i in range(len(res_sh['props']))]

    x = np.arange(len(topics))
    width = 0.35

    plt.figure(figsize=(8,5))
    plt.bar(x - width/2, res_sh['props'], width, label=label_sh)
    plt.bar(x + width/2, res_lz['props'], width, label=label_lz)

    plt.xticks(x, topics)
    plt.ylabel("Proporsi Topik")
    plt.title("Perbandingan Distribusi Topik")
    plt.legend()
    plt.show()
```
